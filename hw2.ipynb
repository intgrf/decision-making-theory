## Задание 1

Для задачи преследования отобразить фигуры на плоскости с помощью инструментального средства или вручную.

Фигура $S_1$ - прямоугольник со сторонами $a$ и $b$.

Фигура $S_2$ - прямоугольник со стороной $c$.


**Центр масс фигуры $S_1$ не принадлежит фигуре $S_2$**

```python
figsize(10, 10)
a = 6
b = 5
c = 3
S1 = Rectangle((2, 4.5), c, c )
S2 = Rectangle((4, 3.5), a, b)
c = PatchCollection([S1, S2], 
                    facecolor=['skyblue', 'lightcoral'],
                    edgecolor=['blue', 'red'],
                    alpha=0.5)

S1_center = ()
fig, ax = plt.subplots(1)
ax.set_xlim([0, 12])
ax.set_ylim([0, 12])

ax.add_collection(c)
plt.plot(3.5, 6, 'k.', markersize=10)
plt.text(3.3, 5.6, r'$O_1$', size=12)
plt.plot(7, 6, 'k.', markersize=10)
plt.text(7, 5.6, r'$O$', size=12)

plt.text(2.2, 4.8, r'$S_1$', size=14)
plt.text(9.5, 3.8, r'$S_2$', size=14)

# доп построение для решения задчи
plt.plot(2, 6, 'k.', markersize=10)
plt.plot([2, 7], [6, 6], 'k-', alpha=0.6)
plt.plot(2, 7.5, 'k.', markersize=10)
plt.plot([2, 7], [7.49, 6], 'k-', alpha=0.6)
plt.plot(4, 6.9, 'k.', markersize=10)
plt.plot([2, 4], [7.49, 6], 'k-', alpha=0.6)
plt.plot(4, 6, 'k.', markersize=10)

plt.text(2, 7.7, r'$A$')
plt.text(4.1, 6.1, r'$B$')
plt.text(4.1, 7, r'$C$')
plt.text(1.7, 5.9, r'$M$')

plt.axis('off')
plt.show()
```
